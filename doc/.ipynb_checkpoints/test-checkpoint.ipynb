```python
def time_merge(dataFrame):
    dataFrame["datetime"] = dataFrame["date"] + " " + dataFrame["time"]
    dataFrame["datetime"] = pd.to_datetime(dataFrame["datetime"])
    dataFrame.drop(["date", "time"], axis=1, inplace=True)
    return dataFrame

gk = pd.read_sql(
    "select date, time, id, pwr from gk_halfhourly_pwr where date>='{}' and date<='{}'".format("2017-06-01", "2018-07-31"), 
    API.AnalysisDB.engine)
gk = time_merge(gk)
gk = gk[gk["id"].apply(lambda x: x[:6]=="XM0001")]
gk.to_csv("./gk.csv", index=None)

dk = pd.read_sql(
    "select date, time, id, pwr from dk_halfhourly_pwr where date>='{}' and date<='{}'".format("2017-06-01", "2018-07-31"),
    API.AnalysisDB.engine)
dk = time_merge(dk)
dk = dk[dk["id"].apply(lambda x: x[:6]=="XM0001")]
dk.to_csv("./dk.csv", index=None)
```

```python
def fillna_time(x):
    x_fillna = pd.DataFrame([(idx, t) for idx in x['id'].unique() for t in pd.date_range(START_DATETIME, END_DATETIME, freq="30min")])
    x_fillna.columns = ["id", "datetime"]
    x_fillna["pwr"] = np.nan
    if x_fillna.shape == x.shape:
        print("No nan Values")
        return x
    else:
        xx = pd.concat([x, x_fillna]).drop_duplicates(["datetime", "id"])
        print("fill {} nan time".format(xx.shape[0]-x.shape[0]))
    return xx
```

```python
from pyod.models import LOF, IForest, KNN, CBLOF

def outlier_detect(x):
    data = x.values.reshape(-1 ,1)
    ol = []
    for n, model in {"lof": LOF(), "iforest": IForest()}.items():
        l = model.fit_predict(data)
        ol.append(l)
    ol = np.all(ol, axis=0).astype(int)
    return ol
```

```python
gk["ol"] = gk.groupby("id")["pwr"].transform(lambda x: outlier_detect(x))
```

```python
def cos_similarity(target, standard):
    target = np.array(target)
    standard = np.array(standard)
    cos = (np.dot(target, standard)) / (np.linalg.norm(target) * np.linalg.norm(standard))
    return cos
```
